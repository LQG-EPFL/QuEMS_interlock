```python
test_inp = interlock.Input(np.random.rand, 'test_in', 'float')

test_out = interlock.Output(np.random.rand, print, 'test_out', 'float')
```

```python
test_inp.add_trigger('if greater than', 0.5)
```

```python
test_inp.check_triggers()
```

```python
inter = interlock.Interlock([test_inp], [test_out])
```

```python
test_out.if_triggered_set_to('waaah!')
test_out.if_normal_set_to('okokok!')
```

```python
inter.run()
```
